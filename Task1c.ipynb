```python
import pandas as pd
data=pd.read_csv("C:/Users/User/OneDrive/Documents/task1.csv")
data
```

```python
print(data.columns)
```

```python
print(data.describe())
```

```python
from sklearn.model_selection import train_test_split
features=['GrLivArea','FullBath','BedroomAbvGr']
x=data[features]
y=data['SalePrice']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
print("MSE= ",mse)
r2=r2_score(y_test,y_pred)
print("R Squared= ",r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

```python
nd=pd.DataFrame({'GrLivArea':[3000],'FullBath':[3],'BedroomAbvGr':[5]})
pp=model.predict(nd)
print("The predicted price of the house is: ",pp[0])
```
